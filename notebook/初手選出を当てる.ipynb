選出はランダムなので、勝ったほうのみを対象にしておく

```python
import pandas as pd

df = pd.read_csv("../data/all_battle_history_season_27.csv")
df.head(2)
```

```python
# 勝ったほうの選出が何だったのかを当てられるような並びにする
# 勝った方のチーム, 負けたほうのチーム, 勝った方の選出順

import ast

features = []

for row in df.iterrows():
    _feature = {}
    player0_name = row[1]["player_0_name"]
    player1_name = row[1]["player_1_name"]
    winner_name = row[1]["winner_name"]

    if winner_name == player0_name:
        team0 = row[1]["player_0_team"]
        team1 = row[1]["player_1_team"]
        team0_first = row[1]["player_0_first"]
        team0_selected = row[1]["player_0_selected"]
    else:
        team0 = row[1]["player_1_team"]
        team1 = row[1]["player_0_team"]
        team0_first = row[1]["player_1_first"]
        team0_selected = row[1]["player_1_selected"]
    
    _feature["team0"] = ast.literal_eval(team0)
    _feature["team1"] = ast.literal_eval(team1)
    _feature["team0_first"] = team0_first
    _feature["team0_selected"] = ast.literal_eval(team0_selected)

    features.append(_feature)

feature_df = pd.DataFrame(features)
```

```python
feature_df
```

```python
# ここで train_test_split しておく
from sklearn.model_selection import train_test_split

train_feature_df, test_feature_df = train_test_split(
    feature_df,
    test_size=0.1,
    random_state=42,
)
train_feature_df, val_feature_df = train_test_split(
    train_feature_df,
    test_size=0.1,
    random_state=42,
)
```

```python
# first 以外を negative として二値予測をするパターン
# feature は team0, team1
# target は team0_first が 1, それ以外は 0

def build_dataset(feature_df):
    dataset = []
    for row in feature_df.iterrows():
        team0 = row[1]["team0"]
        team1 = row[1]["team1"]
        team0_first = row[1]["team0_first"]
        team0_not_selected = list(set(team0) - set([team0_first]))

        dataset.append({
            "team0": team0,
            "team1": team1,
            "team0_first": team0_first,
            "label": 1,
        })

        for not_selected in team0_not_selected:
            dataset.append({
                "team0": team0,
                "team1": team1,
                "team0_first": not_selected,
                "label": 0,
            })

    # dataframe
    dataset_df = pd.DataFrame(dataset)

    return dataset_df

train_df = build_dataset(train_feature_df)
val_df = build_dataset(val_feature_df)
test_df = build_dataset(test_feature_df)
```

```python
from gensim.models import Word2Vec

word2vec_model = Word2Vec.load("../models/word2vec_model.bin")
```

```python
def get_vector(team):
    team_vector = []
    for pokemon in team:
        if pokemon in word2vec_model.wv.key_to_index:
            team_vector.append(word2vec_model.wv[pokemon])
        else:
            print(f"{pokemon} not found in word2vec model.")
    
    if len(team_vector) == 0:
        return None
    
    team_vector = sum(team_vector) / len(team_vector)
    return team_vector

train_df["team0_vector"] = train_df["team0"].apply(get_vector)
train_df["team1_vector"] = train_df["team1"].apply(get_vector)
val_df["team0_vector"] = val_df["team0"].apply(get_vector)
val_df["team1_vector"] = val_df["team1"].apply(get_vector)
test_df["team0_vector"] = test_df["team0"].apply(get_vector)
test_df["team1_vector"] = test_df["team1"].apply(get_vector)

train_df["team0_first_vector"] = train_df["team0_first"].apply(lambda x: word2vec_model.wv[x] if x in word2vec_model.wv.key_to_index else None)
val_df["team0_first_vector"] = val_df["team0_first"].apply(lambda x: word2vec_model.wv[x] if x in word2vec_model.wv.key_to_index else None)
test_df["team0_first_vector"] = test_df["team0_first"].apply(lambda x: word2vec_model.wv[x] if x in word2vec_model.wv.key_to_index else None)

train_df = train_df.dropna()
val_df = val_df.dropna()
test_df = test_df.dropna()
```

```python
# lightgbm dataset に変換

import lightgbm as lgb
import numpy as np

def create_lgb_dataset(df):
    X = []
    y = []
    for row in df.iterrows():
        team0_vector = row[1]["team0_vector"]
        team1_vector = row[1]["team1_vector"]
        team0_first_vector = row[1]["team0_first_vector"]

        # ベクトルを結合
        feature_vector = team0_vector + team1_vector + team0_first_vector
        X.append(feature_vector)
        y.append(row[1]["label"])

    X = np.array(X)
    y = np.array(y)

    # 特徴量の次元数を確認
    print(f"Feature vector shape: {X.shape}")
    # ラベルの数を確認
    print(f"Label shape: {y.shape}")

    return lgb.Dataset(X, label=y)

train_dataset = create_lgb_dataset(train_df)
valid_dataset = create_lgb_dataset(val_df)
test_dataset = create_lgb_dataset(test_df)
```

```python
# 学習

params = {
    'objective': 'binary',
    'metric': 'binary_logloss',
    'boosting_type': 'gbdt',
    'num_leaves': 128,
    'min_data_in_leaf': 20,
    'learning_rate': 0.01,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'max_depth': -1,
}
num_round = 100
verbose_eval = 1

bst = lgb.train(params,
                train_dataset,
                num_round,
                valid_sets=[train_dataset, valid_dataset],
                callbacks=[lgb.early_stopping(stopping_rounds=10, 
                                verbose=True), # early_stopping用コールバック関数
                            lgb.log_evaluation(verbose_eval)] # コマンドライン出力用コールバック関数
                )
```

```python
# test_feature_df に対して予測を行う
import numpy as np

def predict_match(bst, test_feature_df):
    X = []
    pred_first_pokemon = []
    for row in test_feature_df.iterrows():
        team0_vector = row[1]["team0_vector"]
        team1_vector = row[1]["team1_vector"]
        candidate_res = []
        for _candidate in row[1]["team0"]:
            if _candidate in word2vec_model.wv.key_to_index:
                candidate_vector = word2vec_model.wv[_candidate]
                feature_vector = team0_vector + team1_vector + candidate_vector
                candidate_res.append(feature_vector)

        predict_res = bst.predict(candidate_res)
        best_index = np.argmax(predict_res)
        X.append(predict_res)

        pred_first_pokemon.append(row[1]["team0"][best_index])

    X = np.array(X)

    return X, pred_first_pokemon

test_feature_df["team0_vector"] = test_feature_df["team0"].apply(get_vector)
test_feature_df["team1_vector"] = test_feature_df["team1"].apply(get_vector)

y_pred, pred_first_pokemon = predict_match(bst, test_feature_df)
test_feature_df["pred_first_pokemon"] = pred_first_pokemon
```
